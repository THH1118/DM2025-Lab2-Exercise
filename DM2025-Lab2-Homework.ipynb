# **Competition Code**

## 1. Preprocessing Steps

```python
### Add the code related to the preprocessing steps in cells inside this section
!pip install -q sentence-transformers xgboost scikit-learn imbalanced-learn

import re, warnings, math
import numpy as np
import pandas as pd
from collections import Counter

from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import f1_score, confusion_matrix
from sklearn.pipeline import FeatureUnion
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.calibration import CalibratedClassifierCV
from scipy import sparse

from sentence_transformers import SentenceTransformer
import xgboost as xgb
import torch
from imblearn.over_sampling import BorderlineSMOTE

warnings.filterwarnings("ignore")
SEED = 42
np.random.seed(SEED)
print("Using device:", "GPU" if torch.cuda.is_available() else "CPU")

# load data & JSON
emotion_df = pd.read_csv("/content/data/emotion.csv")
split_df   = pd.read_csv("/content/data/data_identification.csv")
df = pd.merge(split_df, emotion_df, on="id", how="left")
train_df = df[df["split"]=="train"].dropna(subset=["emotion"]).copy()
test_df  = df[df["split"]=="test"].copy()

posts_raw = pd.read_json("/content/data/final_posts.json")
posts = pd.json_normalize(posts_raw["root"])
posts = posts[['_source.post.post_id', '_source.post.text']].rename(
    columns={'_source.post.post_id':'id','_source.post.text':'text'}
)
train_df = pd.merge(train_df, posts, on="id", how="left")
test_df  = pd.merge(test_df,  posts, on="id", how="left")

print(f"training data: {len(train_df)} | testing data: {len(test_df)}")

# text cleaning
def clean_text_soft(tx):
    tx = str(tx)
    tx = re.sub(r"http\S+|www\S+"," ", tx)
    tx = re.sub(r"@\w+"," ", tx)
    tx = tx.replace("#", "")
    tx = re.sub(r"[^A-Za-z0-9\s\!\?\.\,;\:\(\)\'\"\-_/]|(?<=\s)[_](?=\s)"," ", tx)  # keep emotional
    tx = re.sub(r"\s+"," ", tx).strip()
    return tx.lower()

def clean_text_hard(tx):
    tx = str(tx)
    tx = re.sub(r"http\S+|www\S+|@\w+|#"," ", tx)
    tx = re.sub(r"[^A-Za-z0-9\s]"," ", tx)
    tx = re.sub(r"\s+"," ", tx).strip()
    return tx.lower()

train_df["clean_soft"] = train_df["text"].apply(clean_text_soft)
train_df["clean_hard"] = train_df["text"].apply(clean_text_hard)
test_df["clean_soft"]  = test_df["text"].apply(clean_text_soft)
test_df["clean_hard"]  = test_df["text"].apply(clean_text_hard)

X_soft_train = train_df["clean_soft"].values
X_hard_train = train_df["clean_hard"].values
y_train      = train_df["emotion"].values
X_soft_test  = test_df["clean_soft"].values
X_hard_test  = test_df["clean_hard"].values

labels       = np.unique(y_train)
L            = len(labels)
label2idx    = {c:i for i,c in enumerate(labels)}
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)
```

## 2. Feature Engineering Steps

```python
### Add the code related to the feature engineering steps in cells inside this section
# TF-IDF feature
tfidf_word = TfidfVectorizer(
    analyzer="word", ngram_range=(1,2),
    max_features=22000, min_df=2, max_df=0.995, sublinear_tf=True
)
tfidf_char = TfidfVectorizer(
    analyzer="char", ngram_range=(2,6),
    min_df=2, sublinear_tf=True
)
tfidf_union = FeatureUnion([("word", tfidf_word), ("char", tfidf_char)])
X_all_tfidf  = tfidf_union.fit_transform(X_soft_train)
X_test_tfidf = tfidf_union.transform(X_soft_test)

# Count feature
cv_word = CountVectorizer(analyzer="word", ngram_range=(1,2), min_df=2, max_df=0.995)
cv_char = CountVectorizer(analyzer="char", ngram_range=(2,6), min_df=2)
X_cnt_word = cv_word.fit_transform(X_soft_train)
X_cnt_char = cv_char.fit_transform(X_soft_train)
X_cnt      = sparse.hstack([X_cnt_word, X_cnt_char]).tocsr()
X_cnt_test = sparse.hstack([cv_word.transform(X_soft_test), cv_char.transform(X_soft_test)]).tocsr()

le_full    = LabelEncoder().fit(y_train)
y_enc_full = le_full.transform(y_train)
C = len(le_full.classes_)

def log_count_ratio(X, y_enc, c):
    p = 1 + X[y_enc==c].sum(0)
    q = 1 + X[y_enc!=c].sum(0)
    r = np.log(p / q)
    return np.asarray(r).ravel()

# SBERT
def encode_sbert(encoder_name, texts):
    model = SentenceTransformer(encoder_name, device='cuda' if torch.cuda.is_available() else 'cpu')
    emb = model.encode(list(texts), show_progress_bar=True, batch_size=64)
    return emb

# two encoder
encoders = [
    'all-mpnet-base-v2',
    'paraphrase-mpnet-base-v2',
]

# build SBERT vector（soft / hard ）
sbert_soft = {}
sbert_hard = {}
for name in encoders:
    print(f"\n Encoding ({name}) on SOFT text ...")
    sbert_soft[name] = encode_sbert(name, X_soft_train), encode_sbert(name, X_soft_test)
    print(f" Encoding ({name}) on HARD text ...")
    sbert_hard[name] = encode_sbert(name, X_hard_train), encode_sbert(name, X_hard_test)

# XGBoost
le_xgb = LabelEncoder().fit(y_train)
y_train_enc = le_xgb.transform(y_train)

def effective_num_weights(y, num_classes, beta=0.999):
    binc = np.bincount(y, minlength=num_classes).astype(float)
    eff_num = 1.0 - np.power(beta, binc)
    cls_w = (1.0 - beta) / np.clip(eff_num, 1e-8, None)
    cls_w = cls_w / cls_w.mean()
    return cls_w
```

## 3. Model Implementation Steps

```python
### Add the code related to the model implementation steps in cells inside this section
# A：TF-IDF + LinearSVC
oof_A = np.zeros((len(X_soft_train), L))
test_A_folds = []
for tr_idx, va_idx in skf.split(X_all_tfidf, y_train):
    X_tr, X_va = X_all_tfidf[tr_idx], X_all_tfidf[va_idx]
    y_tr, y_va = y_train[tr_idx], y_train[va_idx]
    base = LinearSVC(C=1.0, class_weight="balanced", random_state=SEED)
    clfA = CalibratedClassifierCV(base, method="isotonic", cv=3)
    clfA.fit(X_tr, y_tr)
    oof_A[va_idx] = clfA.predict_proba(X_va)
    test_A_folds.append(clfA.predict_proba(X_test_tfidf))
P_test_A   = np.mean(test_A_folds, axis=0)
oof_pred_A = labels[oof_A.argmax(1)]
print("[A] OOF macro-F1 =", f1_score(y_train, oof_pred_A, average="macro"))

# B：TF-IDF + LogisticRegression
oof_B = np.zeros((len(X_soft_train), L))
test_B_folds = []
for tr_idx, va_idx in skf.split(X_all_tfidf, y_train):
    X_tr, X_va = X_all_tfidf[tr_idx], X_all_tfidf[va_idx]
    y_tr, y_va = y_train[tr_idx], y_train[va_idx]
    baseLR = LogisticRegression(
        solver="saga", penalty="l2", C=2.0, max_iter=200,
        multi_class="multinomial", class_weight="balanced", n_jobs=-1, random_state=SEED
    )
    clfB = CalibratedClassifierCV(baseLR, method="isotonic", cv=3)
    clfB.fit(X_tr, y_tr)
    oof_B[va_idx] = clfB.predict_proba(X_va)
    test_B_folds.append(clfB.predict_proba(X_test_tfidf))
P_test_B   = np.mean(test_B_folds, axis=0)
oof_pred_B = labels[oof_B.argmax(1)]
print("[B] OOF macro-F1 =", f1_score(y_train, oof_pred_B, average="macro"))

# C：NB-SVM
oof_C = np.zeros((len(X_soft_train), C))
test_C_folds = []
for tr_idx, va_idx in skf.split(X_cnt, y_enc_full):
    X_tr, X_va = X_cnt[tr_idx], X_cnt[va_idx]
    y_tr, y_va = y_enc_full[tr_idx], y_enc_full[va_idx]
    preds_val = np.zeros((len(va_idx), C))
    preds_tst = np.zeros((X_cnt_test.shape[0], C))
    for c in range(C):
        r_c = log_count_ratio(X_tr, y_tr, c)
        X_tr_rc = X_tr.multiply(r_c)
        X_va_rc = X_va.multiply(r_c)
        X_te_rc = X_cnt_test.multiply(r_c)
        y_bin = (y_tr==c).astype(int)
        pos = int(y_bin.sum()); neg = len(y_bin)-pos
        w_pos = float(np.sqrt(neg/max(pos,1)))
        svc = LinearSVC(C=1.0, class_weight={1:w_pos, 0:1.0}, random_state=SEED)
        svc.fit(X_tr_rc, y_bin)
        val_score = svc.decision_function(X_va_rc)
        tst_score = svc.decision_function(X_te_rc)
        preds_val[:, c] = 1/(1+np.exp(-val_score))
        preds_tst[:, c] = 1/(1+np.exp(-tst_score))
    preds_val = (preds_val.T / preds_val.sum(1)).T
    preds_tst = (preds_tst.T / preds_tst.sum(1)).T
    oof_C[va_idx] = preds_val
    test_C_folds.append(preds_tst)

idx_map_C = [np.where(le_full.classes_==c)[0][0] for c in labels]
oof_C     = oof_C[:, idx_map_C]
P_test_C  = np.mean(test_C_folds, axis=0)[:, idx_map_C]
oof_pred_C= labels[oof_C.argmax(1)]
print("[C] OOF macro-F1 =", f1_score(y_train, oof_pred_C, average="macro"))

# D：SBERT(+XGB) with TTApnet
name = 'all-mpnet-base-v2'
X_soft_tr, X_soft_te = sbert_soft[name]
X_hard_tr, X_hard_te = sbert_hard[name]

oof_D = np.zeros((len(X_soft_train), L))
test_D_folds = []
skf2 = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)
for tr_idx, va_idx in skf2.split(X_soft_tr, y_train_enc):
    X_tr, X_va = X_soft_tr[tr_idx], X_soft_tr[va_idx]
    y_tr, y_va = y_train_enc[tr_idx], y_train_enc[va_idx]

    # BorderlineSMOTE + Effective-number
    cnt = Counter(y_tr); maj = max(cnt.values())
    sampling_strategy = {c: max(v, int(0.55*maj)) for c, v in cnt.items()}
    sm = BorderlineSMOTE(k_neighbors=3, sampling_strategy=sampling_strategy, random_state=SEED)
    X_tr_sm, y_tr_sm = sm.fit_resample(X_tr, y_tr)
    cls_w = effective_num_weights(y_tr_sm, L, beta=0.999)
    sw_tr = cls_w[y_tr_sm]

    xgbD = xgb.XGBClassifier(
        objective='multi:softprob', num_class=L,
        learning_rate=0.10, max_depth=6, n_estimators=420,
        subsample=0.9, colsample_bytree=0.95,
        eval_metric='mlogloss', tree_method='hist',
        reg_alpha=0.0, reg_lambda=1.0, random_state=SEED
    )
    xgbD.fit(X_tr_sm, y_tr_sm, sample_weight=sw_tr)
    proba_va_soft = xgbD.predict_proba(X_soft_tr[va_idx])
    proba_va_hard = xgbD.predict_proba(X_hard_tr[va_idx])
    proba_va = (proba_va_soft + proba_va_hard) / 2.0

    proba_te_soft = xgbD.predict_proba(X_soft_te)
    proba_te_hard = xgbD.predict_proba(X_hard_te)
    proba_te = (proba_te_soft + proba_te_hard) / 2.0

    idx_map_D = [np.where(le_xgb.classes_==c)[0][0] for c in labels]
    oof_D[va_idx] = proba_va[:, idx_map_D]
    test_D_folds.append(proba_te[:, idx_map_D])

P_test_D   = np.mean(test_D_folds, axis=0)
oof_pred_D = labels[oof_D.argmax(1)]
print("[D] OOF macro-F1 =", f1_score(y_train, oof_pred_D, average="macro"))

# E：SBERT(+CalibSVC) with TTA（soft/hard）
oof_E = np.zeros((len(X_soft_train), L))
test_E_folds = []
for tr_idx, va_idx in skf2.split(X_soft_tr, y_train):
    X_tr, y_tr = X_soft_tr[tr_idx], y_train[tr_idx]
    base = LinearSVC(C=1.0, class_weight="balanced", random_state=SEED)
    clfE = CalibratedClassifierCV(base, method="isotonic", cv=3)
    clfE.fit(X_tr, y_tr)

    # TTA
    p_va_soft  = clfE.predict_proba(X_soft_tr[va_idx])
    p_va_hard  = clfE.predict_proba(X_hard_tr[va_idx])
    p_te_soft  = clfE.predict_proba(X_soft_te)
    p_te_hard  = clfE.predict_proba(X_hard_te)

    oof_E[va_idx] = (p_va_soft + p_va_hard)/2.0
    test_E_folds.append((p_te_soft + p_te_hard)/2.0)

P_test_E   = np.mean(test_E_folds, axis=0)
oof_pred_E = labels[oof_E.argmax(1)]
print("[E] OOF macro-F1 =", f1_score(y_train, oof_pred_E, average="macro"))

# F：SBERT(+XGB) with TTA（soft/hard） for paraphrase-mpnet
name2 = 'paraphrase-mpnet-base-v2'
X2_soft_tr, X2_soft_te = sbert_soft[name2]
X2_hard_tr, X2_hard_te = sbert_hard[name2]

oof_F = np.zeros((len(X_soft_train), L))
test_F_folds = []
for tr_idx, va_idx in skf2.split(X2_soft_tr, y_train_enc):
    X_tr, X_va = X2_soft_tr[tr_idx], X2_soft_tr[va_idx]
    y_tr, y_va = y_train_enc[tr_idx], y_train_enc[va_idx]

    cnt = Counter(y_tr); maj = max(cnt.values())
    sampling_strategy = {c: max(v, int(0.55*maj)) for c, v in cnt.items()}
    sm = BorderlineSMOTE(k_neighbors=3, sampling_strategy=sampling_strategy, random_state=SEED)
    X_tr_sm, y_tr_sm = sm.fit_resample(X_tr, y_tr)
    cls_w = effective_num_weights(y_tr_sm, L, beta=0.999)
    sw_tr = cls_w[y_tr_sm]

    xgbF = xgb.XGBClassifier(
        objective='multi:softprob', num_class=L,
        learning_rate=0.10, max_depth=6, n_estimators=420,
        subsample=0.9, colsample_bytree=0.95,
        eval_metric='mlogloss', tree_method='hist',
        reg_alpha=0.0, reg_lambda=1.0, random_state=SEED
    )
    xgbF.fit(X_tr_sm, y_tr_sm, sample_weight=sw_tr)

    proba_va = (xgbF.predict_proba(X2_soft_tr[va_idx]) + xgbF.predict_proba(X2_hard_tr[va_idx])) / 2.0
    proba_te = (xgbF.predict_proba(X2_soft_te) + xgbF.predict_proba(X2_hard_te)) / 2.0

    idx_map_F = [np.where(le_xgb.classes_==c)[0][0] for c in labels]
    oof_F[va_idx] = proba_va[:, idx_map_F]
    test_F_folds.append(proba_te[:, idx_map_F])

P_test_F   = np.mean(test_F_folds, axis=0)
oof_pred_F = labels[oof_F.argmax(1)]
print("[F SBERT(paraphrase-mpnet)+XGB+TTA] OOF macro-F1 =", f1_score(y_train, oof_pred_F, average="macro"))

# Stacking（A/B/C/D/E/F）+ taus + Per-class Logit Bias + Pairwise flip
P_oof = np.hstack([oof_A, oof_B, oof_C, oof_D, oof_E, oof_F])   # [N, 6*L]
P_tst = np.hstack([P_test_A, P_test_B, P_test_C, P_test_D, P_test_E, P_test_F])

blender = LogisticRegression(
    solver="lbfgs", multi_class="multinomial",
    class_weight="balanced", max_iter=400, random_state=SEED
)
blender.fit(P_oof, y_train)
proba_oof_bl = blender.predict_proba(P_oof)
base_pred    = proba_oof_bl.argmax(1)
print("[STACK(raw)] OOF macro-F1 =", f1_score(y_train, labels[base_pred], average="macro"))

taus = np.ones(L)
best_f1 = f1_score(y_train, labels[base_pred], average="macro")
for j in range(L):
    best_tau_j, best_f1_j = 1.0, best_f1
    for t in np.linspace(0.6, 1.8, 13):
        P = proba_oof_bl.copy()
        P[:, j] = P[:, j] ** (1.0 / t)
        P = (P.T / P.sum(1)).T
        pred = P.argmax(1)
        f1 = f1_score(y_train, labels[pred], average="macro")
        if f1 > best_f1_j:
            best_f1_j, best_tau_j = f1, t
    taus[j] = best_tau_j
    proba_oof_bl[:, j] = proba_oof_bl[:, j] ** (1.0 / best_tau_j)
    proba_oof_bl = (proba_oof_bl.T / proba_oof_bl.sum(1)).T
    best_f1 = best_f1_j

print("Best taus per class:", dict(zip(labels, taus)))
print("[STACK+Temp] OOF macro-F1 =", best_f1)

#Per-class Logit Bias
eps = 1e-9
def prob_to_logits(P):
    return np.log(np.clip(P, eps, None))

logits_oof = prob_to_logits(proba_oof_bl)
b = np.zeros(L, dtype=float)
best_f1_bias = best_f1

grid = np.linspace(-0.8, 0.8, 17)
for it in range(2):
    improved = False
    for j in range(L):
        best_bj, best_local = b[j], best_f1_bias
        for bj in grid:
            b_try = b.copy()
            b_try[j] = bj
            z = logits_oof + b_try
            P = np.exp(z); P = (P.T / P.sum(1)).T
            pred = P.argmax(1)
            f1 = f1_score(y_train, labels[pred], average="macro")
            if f1 > best_local:
                best_local, best_bj = f1, bj
        if best_bj != b[j]:
            b[j] = best_bj
            best_f1_bias = best_local
            improved = True
    if not improved:
        break

print("Per-class logit bias b:", dict(zip(labels, b)))
print("[STACK+Temp+Bias] OOF macro-F1 =", best_f1_bias)

#Pairwise margin flip
P_oof_bias = np.exp(logits_oof + b); P_oof_bias = (P_oof_bias.T / P_oof_bias.sum(1)).T
pred_labels = labels[P_oof_bias.argmax(1)]
cm = confusion_matrix(y_train, pred_labels, labels=labels)
pairs = []
for i in range(L):
    for j in range(L):
        if i!=j:
            pairs.append(((labels[i], labels[j]), cm[i,j] + cm[j,i]))
pairs = [p[0] for p in sorted(pairs, key=lambda x: x[1], reverse=True)[:4]]

pair_tau = {}
P_curr = P_oof_bias.copy()
for (A,B) in pairs:
    ia, ib = label2idx[A], label2idx[B]
    best_pair_f1, best_tau = -1, 1.0
    for t in np.linspace(0.8, 1.4, 13):
        pred = P_curr.argmax(1)
        pa = P_curr[:, ia]; pb = P_curr[:, ib]
        mask = (pred==ia) & (pb/np.clip(pa,1e-8,None) > t)
        pred_adj = pred.copy()
        pred_adj[mask] = ib
        f1 = f1_score(y_train, labels[pred_adj], average="macro")
        if f1 > best_pair_f1:
            best_pair_f1, best_tau = f1, t
    pair_tau[(A,B)] = best_tau

print("Top confused pairs & τ:", pair_tau)

# test
proba_tst_bl = blender.predict_proba(P_tst)
for j in range(L):
    proba_tst_bl[:, j] = proba_tst_bl[:, j] ** (1.0 / taus[j])
proba_tst_bl = (proba_tst_bl.T / np.clip(proba_tst_bl.sum(1), 1e-8, None)).T

logits_tst = np.log(np.clip(proba_tst_bl, 1e-9, None))
P_tst_bias = np.exp(logits_tst + b); P_tst_bias = (P_tst_bias.T / P_tst_bias.sum(1)).T
pred_tst = P_tst_bias.argmax(1)

# Pairwise flip
for (A,B), tau in pair_tau.items():
    ia, ib = label2idx[A], label2idx[B]
    pa = P_tst_bias[:, ia]; pb = P_tst_bias[:, ib]
    mask = (pred_tst==ia) & (pb/np.clip(pa,1e-8,None) > tau)
    pred_tst[mask] = ib

test_pred = labels[pred_tst]
submission = pd.DataFrame({"id": test_df["id"], "emotion": test_pred})
submission.to_csv("submission_final_v3.csv", index=False)
print("\nsubmission_final_v3.csv ok ！")
```
